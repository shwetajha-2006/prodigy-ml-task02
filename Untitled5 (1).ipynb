```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('archive.zip')
df.head()
```

```python
df.info()
df.describe()
sns.pairplot(df)
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, random_state=42)
  kmeans.fit(X_scaled)
  wcss.append(kmeans.inertia_)

# Plot the elbow method after the loop
plt.plot(range(1,11), wcss)
plt.title('Elbow Method') # Setting the title once
plt.xlabel('Number of Clusters') # Adding x-axis label for clarity
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
df['Cluster'] = clusters
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=df, palette='viridis', s=100) # Corrected palette name
plt.title('Customer Segment using KMeans')
plt.show()
```
